```python
import numpy as np
import matplotlib.pyplot as plt

def show_help():
    print("Інструкція до функції:")
    print("1. Введіть коефіцієнти a, b та c для квадратного рівняння у формі ax^2 + bx + c = 0.")
    print("2. Функція поверне розв'язки цього рівняння у вигляді numpy масиву.")
    print("3. Щоб завершити роботу функції, введіть 'exit'.")
def input_coefficient(prompt):
    while True:
        val = input(prompt)
        if val == "help":
            show_help()
        elif val == "exit":
            return None
        else:
            try:
                val = float(val)
                if val == 0 and prompt.startswith("Введіть коефіцієнт a: "):
                    print("Коефіцієнт a не може дорівнювати нулю. Будь ласка, введіть коректне значення.")
                else:
                    return val
            except ValueError:
                print("Будь ласка, введіть коректний числовий коефіцієнт, exit або help.")

def solve_quadratic():
    while True:
        a = input_coefficient("Введіть коефіцієнт a, help, або exit: ")
        if a is None:
            return

        b = input_coefficient("Введіть коефіцієнт b: ")
        if b is None:
            return

        c = input_coefficient("Введіть коефіцієнт c: ")
        if c is None:
            return

        discriminant = b**2 - 4*a*c
        x1 = None
        x2 = None

        if discriminant > 0:
            x1 = (-b + np.sqrt(discriminant)) / (2*a)
            x2 = (-b - np.sqrt(discriminant)) / (2*a)
        elif discriminant == 0:
            x1 = -b / (2*a)

        solutions = [x for x in [x1, x2] if x is not None]

        if not solutions:
            print("Рівняння не має дійсних розв'язків.")
        else:
            # Побудова графіку квадратичної функції та позначення розв'язків на графіку
            x = np.linspace(min(solutions) - 1, max(solutions) + 1, 400)
            y = a * x**2 + b * x + c

            plt.figure()
            plt.plot(x, y, label=f'{a}x^2 + {b}x + {c}')
            plt.xlabel('x')
            plt.ylabel('y')
            plt.grid(True)

            for solution in solutions:
                plt.plot(solution, a * solution**2 + b * solution + c, 'ro', label=f'Розв\'язок: x = {solution}')

            plt.legend()
            plt.title('Графік квадратичної функції')
            plt.show()

solutions = solve_quadratic()
```
